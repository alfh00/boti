```python
import ccxt
from datetime import datetime, timedelta
import pandas as pd
import time
from multiprocessing import Pool, cpu_count

# Initialisation de l'API Bitget avec CCXT
exchange = ccxt.bitget()

# Paramètres pour récupérer les données
symbol = 'CRV/USDT'
timeframe = '1m'  # Période de la bougie (1m, 5m, 1h, etc.)
start_date = exchange.parse8601('2024-07-01T00:00:00Z')  # Date de début en format ISO 8601
end_date = exchange.parse8601(datetime.utcnow().isoformat() + 'Z')  # Date actuelle

# Fonction pour télécharger un segment de bougies
def fetch_candles(segment):
    since, _ = segment
    try:
        candles = exchange.fetch_ohlcv(symbol, timeframe, since, limit=1000)
        return candles
    except ccxt.BaseError as e:
        print(f"Erreur lors du téléchargement des bougies : {str(e)}")
        return []

# Diviser la période totale en segments de 1000 bougies
def create_segments(start, end, interval=1000):
    segments = []
    while start < end:
        segments.append((start, min(start + (interval * 60 * 1000) - 1, end)))
        start += interval * 60 * 1000
    return segments

segments = create_segments(start_date, end_date)

# Utilisation du multiprocessing pour accélérer le téléchargement
if __name__ == '__main__':
    with Pool(processes=cpu_count()) as pool:
        all_candles = pool.map(fetch_candles, segments)

    # Flatten the list of lists
    all_candles = [item for sublist in all_candles for item in sublist]
    all_candles 

    # Convertir les données en DataFrame Pandas pour l'analyse
    # df = pd.DataFrame(all_candles, columns=['timestamp', 'open', 'high', 'low', 'close', 'volume'])
    # df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')
    # df.set_index('timestamp', inplace=True)

    # # Sauvegarde des données dans un fichier CSV
    # df.to_csv('aaa.csv')

    # print("Données téléchargées et sauvegardées avec succès.")
```
